```python
 #graphic card
import re # Import the 're' module for regular expressions
def extract_brand(description):
    # Define a list of known GPU brands
    brands = ['NVIDIA', 'Intel', 'AMD', 'Qualcomm','ARM','Apple',]
    # Create a regex pattern to search for these brands
    pattern = '|'.join(brands)
    # Search for the pattern in the description
    match = re.search(pattern, description, re.IGNORECASE)
    # Return the matched brand or 'Unknown' if no match is found
    return match.group(0) if match else 'Unknown'

gpu_brand = df1['graphic_card'].apply(extract_brand) # Apply the function to the 'graphic_card' column of df1
```

```python
# Function to classify GPU
def classify_gpu(gpu_name):
    integrated_keywords = ['RTX', 'RX', 'Arc A-series', 'NVIDIA']
    if any(keyword.lower() in gpu_name.lower() for keyword in integrated_keywords):
        return 'Dedicated'
    else:
        return 'Integrated'

# Apply the function to the DataFrame
gpu_type = df1['graphic_card'].apply(classify_gpu)
```
